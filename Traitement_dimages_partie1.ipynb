```python
img_c = PI.open('/home/depinfo/Documents/Python/couleurs.png')

def decoder(image):
    
    colonne , ligne=image.size
    
    for i in range(colonne):
        for j in range(ligne):
            (r,g,b) = image.getpixel((i,j))
            der_bits_r = valeur_derniers_bits(r)
            der_bits_r = decaler(der_bits_r)
            
            der_bits_g = valeur_derniers_bits(g)
            der_bits_g = decaler(der_bits_g)
            
            der_bits_b = valeur_derniers_bits(b)
            der_bits_b = decaler(der_bits_b)

            x=(der_bits_r, der_bits_g, der_bits_b)
            image.putpixel((i,j),x)
            
    return image 
            
decoder(img_c)    
```

```python
img = PI.open('/home/depinfo/Documents/Python/index.png')
img_c = PI.open('/home/depinfo/Documents/Python/couleurs.png')

def valeurs_premiers_bits(n,v):
    
    n_bin = np.binary_repr(n, width=8)
    return n_bin[:v]

def cache(image_illusion,image_cacher):
    colonne , ligne=image_illusion.size
    
    for i in range(colonne):
        for j in range(ligne):
            (r1,g1,b1) = image_illusion.getpixel((i,j))
            (r2,g2,b2) = image_cacher.getpixel((i,j))
            
            r1 = valeurs_premiers_bits(r1,5)
            r1 = r1 + valeurs_premiers_bits(r2,3)
            r = int(r1,2)
            
            g1 = valeurs_premiers_bits(g1,5)
            g1 = g1 + valeurs_premiers_bits(g2,3)
            g = int(g1,2)
            
            b1 = valeurs_premiers_bits(b1,5)
            b1 = g1 + valeurs_premiers_bits(b2,3)
            b = int(b1,2)
            
            image_illusion.putpixel((i,j),(r,g,b))
            
            
    return image_illusion

cache(img_c,img)
            
```
