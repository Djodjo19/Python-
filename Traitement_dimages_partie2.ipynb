```python
from PIL import Image

def photomaton(image):
    h,l=image.size
    new_image = PI.new('RGB',image.size)
    
    for i in range(h):
        for j in range(l):
            a = image.getpixel((i,j))
            
            if (i%2==0 and j%2==0):
                (x,y) = (i//2,j//2)
            if (i%2==0 and j%2!=0):
                (x,y) = (i//2,(j+l)//2)
            if (i%2!=0 and j%2==0):
                (x,y) = ((i+h)//2,j//2)
            if (i%2!=0 and j%2!=0):
                (x,y) = ((i+h)//2,(j+l)//2)
            
            new_image.putpixel((x,y), a)
    return new_image

photomaton(open('/home/depinfo/Documents/Python/joconde.png'))
                
    
            
```
